## 2nd Part

### Convert data with latitude and longitude of Toronto's Postal Codes into a dataframe

```python
csv_file = 'https://cocl.us/Geospatial_data'
geospatial_df = pd.read_csv(csv_file)

geospatial_df.head()
```

```python
geospatial_df.shape
```

### Rename the Postal Code feature as PostalCode to match with the same feature in toronto_df

```python
geospatial_df.rename(columns={'Postal Code':'PostalCode'}, inplace=True)
geospatial_df.head()
```

### Merge both dataframes on PostalCode

```python
toronto_merged = pd.merge(toronto_df, geospatial_df, on=['PostalCode'])
toronto_merged.head()
```

```python
toronto_merged.shape
```

## 3rd Part

### Cluster Neighborhoods

### Import the libraries needed and create a map of Toronto with the boroughs and neighborhoods

```python
import matplotlib.cm as cm
import matplotlib.colors as colors

import folium
```

```python
map_toronto = folium.Map(location=[43.651070, -79.347015], zoom_start=10)

for lat, lng, borough, neighborhood in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Borough'], toronto_merged['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)
    
map_toronto
```

### Import library for clusters, create the clusters and visualize them in the map

```python
from sklearn.cluster import KMeans
```

```python
kclusters = 5

toronto_merged_clustering = toronto_merged.drop(['PostalCode', 'Borough', 'Neighborhood'], 1)
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_merged_clustering)

kmeans.labels_

toronto_merged.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
toronto_merged.head()
```

```python
map_clusters = folium.Map(location=[43.651070, -79.347015], zoom_start=10)

x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

markers_colors = []
for lat, lng, neighborhood, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
    
map_clusters
```
